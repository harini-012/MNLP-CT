```python
from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
text=input()
#Convert to lower case
text=text.lower()
#removing punctuations
for t in string.punctuation:
    text=text.replace(t,"")
#tokenization
words=word_tokenize(text)
print("Tokens:\n",words)
#total words
total_words=len(words)
print(total_words)
#vocabulary size
unique_words=set(words)
print("Vocabulary size:",len(unique_words))
#word frequency
freq=Counter(words)
print("\n Word Frequency:",freq)
print("\n Top 10 words")
for word,count in freq.most_common(10):
    print(word,":",count)
#TypeTokenRatio
ttr=len(unique_words)/total_words
print("\n Type Token Ratio:",round(ttr,2))
#Remove the stop words
stop_words=set(stopwords.words('english'))
hapax_legomena=[]
for word,count in freq.items():
    if count==1:
        hapax_legomena.append(word)
print("\nRare words:",happax_legomena)
filtered_words=[w for w in words if w not in stop_words]
print("\nFiltered words:",filtered_words)
#Frequency after stop word removal
filtered_freq=Counter(filtered_words)
print("Filtered Words Frequency Count:",filtered_freq)
#Chart Visualization
word_list=list(filtered_freq.keys())
counts_list=list(filtered_freq.values())
plt.figure(figsize=(8,5))
plt.bar(word_list,counts_list)
plt.title('NLP Word Frequency analysis')
plt.xlabel('Word List')
plt.ylabel('Freq List')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
#Next word prediction using N-Gram
from collections import defaultdict,Counter
#training corpus 
text="""
What is AI
What is GenAI
What is LLM
What is RAG
What is AI
What is tokenization"""
#tokenization
words=text.lower().split()
#build bigram
model=defaultdict(list)
for i in range(len(words)-2):
    w1,w2,w3=words[i],words[i+1],words[i+2]
    model[(w1,w2)].append(w3)
#input phrase
input_words=("what","is")
#predict
if input_words in model:
    prediction=Counter(model[input_words]).most_common(1)[0][0]
    print("Input: What is")
    print("NExt word prediction:",prediction)
else:
    print("No words is predicted")
```
